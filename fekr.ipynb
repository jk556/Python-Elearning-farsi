```python
def getInput():
    a = input()
    b = list(a)
    hads = []
    for letter in b:
        hads.append(int(letter))
    return hads    
```

```python
cc = 0
cp = 0
cnt = 1
Max_hads = 5
win = 0
arr = [1,2,3,4,5,6]
ramz = random.permutation(arr)
ramz = ramz[2:]
while cnt <= Max_hads:
    cc = 0
    cp = 0
    hads = getInput()
    for i in range(0,4):
        if ramz[i] == hads[i]:
            cp += 1
    for i in range(0,4):
        for j in range(0,4):
            if hads[i] == ramz[j]:
                cc += 1
                break
    cc = cc - cp
    cnt += 1
    if (cp == 4):
        print("You Win!!!")
        win = 1
        break
    else:
        print(f'cc is {cc} and cp is {cp}\n')
if (win == 0):
    print("Game Over!\n Ramz :",*ramz,ramz)
```
